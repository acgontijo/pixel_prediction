### Accessing the data 

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import pkg_resources
from multiearth_challenge import tiff_file_tools as tft
import os
import random
```

```python
directory_y = '../data/raw/targets'
substring = '2016_08'
num_files_to_select = 1000
```

```python
filtered_filenames_y = [f for f in os.listdir(directory_y) if substring in f]
filtered_filenames_y
```

```python
len(filtered_filenames_y)
```

```python
selected_filenames_y = random.sample(filtered_filenames_y, num_files_to_select)
selected_filenames_y
```

```python
selected_filepaths_y = [os.path.join(directory_y, filename) for filename in selected_filenames_y]
selected_filepaths_y
```

```python
directory_x = '../data/raw/images'
```

```python
# Build a dictionary to store matching substrings in directory_x
substring_dict_x = {}

# Preprocess the files in directory_x and store them by substring
for filename in os.listdir(directory_x):
    # Extract the substring that you're looking for
    matching_substring = filename[-27:-8]
    
    # Only store files that correspond to RGB bands (SR_B2, SR_B3, SR_B4)
    if 'SR_B4' in filename:   # SR_B4 = Red
        band = 'R'
    elif 'SR_B3' in filename:  # SR_B3 = Green
        band = 'G'
    elif 'SR_B2' in filename:  # SR_B2 = Blue
        band = 'B'
    else:
        # Skip files that are not part of the RGB bands
        continue
    
    # Add the file to the dictionary
    if matching_substring not in substring_dict_x:
        substring_dict_x[matching_substring] = {'R': None, 'G': None, 'B': None}
    
    # Store filenames by band (only store one file for each band per matching substring)
    substring_dict_x[matching_substring][band] = filename

# Now, for each selected filename, find corresponding RGB files in directory_x
selected_filepaths_x = []

for selected_filename in selected_filenames_y:
    matching_substring = selected_filename[-27:-8]
    
    # Check if the substring exists in the preprocessed dictionary
    if matching_substring in substring_dict_x:
        corresponding_files = substring_dict_x[matching_substring]
        
        # Only add files that have all three RGB bands (R, G, and B)
        if all(corresponding_files[band] is not None for band in ['R', 'G', 'B']):
            # Append the full file paths to the result list
            selected_filepaths_x.append(os.path.join(directory_x, corresponding_files['R']))
            selected_filepaths_x.append(os.path.join(directory_x, corresponding_files['G']))
            selected_filepaths_x.append(os.path.join(directory_x, corresponding_files['B']))
```

```python
# Function to load the image and convert it into a numpy array
def load_image(image_path):
    image = Image.open(image_path)  # Open image using PIL
    return np.array(image)  # Convert to numpy array

# To store the images as numpy arrays
x_images = []  # Will hold RGB stacked images
y_images = []  # Will hold single images

# Process the y images (one per entry)
for filepath_y in selected_filepaths_y:
    image_y = load_image(filepath_y)
    y_images.append(image_y)

# Process the x images (RGB stacks)
for i in range(0, len(selected_filepaths_x), 3):  # Step through the list in steps of 3
    r_image_path = selected_filepaths_x[i]     # Red image (SR_B4)
    g_image_path = selected_filepaths_x[i+1]   # Green image (SR_B3)
    b_image_path = selected_filepaths_x[i+2]   # Blue image (SR_B2)
    
    # Load each band and convert to numpy arrays
    r_image = load_image(r_image_path)
    g_image = load_image(g_image_path)
    b_image = load_image(b_image_path)
    
    # Stack the three channels along the third axis (channel axis)
    rgb_image = np.stack([r_image, g_image, b_image], axis=-1)
    
    # Append the stacked RGB image to the list
    x_images.append(rgb_image)

# Convert the lists of images to numpy arrays (if needed as a batch)
x_images = np.array(x_images)
y_images = np.array(y_images)
```

```python
from PIL import Image
import numpy as np

# Resize using PIL
y_images_resized = [np.array(Image.fromarray(image).resize((85, 85))) for image in y_images]
y_images_resized[0]
```

```python
# Normalize the RGB images in x_images to the range [0, 255]
def normalize_image(image):
    # Normalize to [0, 255]
    image = image.astype(np.float32)  # Convert to float for scaling
    max_value = np.max(image)
    if max_value > 255:
        image = (image / max_value) * 255  # Scale to 0-255 range
    image = np.clip(image, 0, 255)  # Ensure values are in [0, 255]
    return image.astype(np.uint8)  # Convert back to uint8 for image display

# Normalize all x_images
x_images_normalized = np.array([normalize_image(image) for image in x_images])
```

```python
import matplotlib.pyplot as plt
import random

# Select a random index to plot (you can select any index or use a fixed index)
random_idx = random.choice(range(len(x_images_normalized)))

# Get the corresponding normalized RGB image from x_images_normalized
image_to_plot_x = x_images_normalized[random_idx]

# Get the corresponding single-channel image from y_images
image_to_plot_y = y_images_resized[random_idx]

# Create a figure with 1 row and 2 columns to plot side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot the normalized RGB image (x)
axes[0].imshow(image_to_plot_x)
axes[0].set_title(f'Normalized RGB Image {random_idx}')
axes[0].axis('off')  # Hide axis for better visualization

# Plot the single-channel grayscale image (y)
axes[1].imshow(image_to_plot_y, cmap='gray')
axes[1].set_title(f'Grayscale Image {random_idx}')
axes[1].axis('off')  # Hide axis for better visualization

# Show the plot
plt.tight_layout()
plt.show()
```
