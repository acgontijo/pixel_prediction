```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Assuming x_train and y_train are already defined
# x_train = ... (Your RGB images)
# y_train = ... (Your y images)

# Pick a random index
random_index = random.choice(range(len(X_test)))

# Retrieve the corresponding images
random_x_image = X_test[random_index]
random_y_image = y_test[random_index]

# Plot the images side by side
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot the x image (RGB)
axes[0].imshow(random_x_image)
axes[0].set_title("Random X (RGB)")
axes[0].axis('off')  # Hide axes for better visualization

# Plot the y image (single)
axes[1].imshow(random_y_image, cmap='gray')
axes[1].set_title("Matching Y")
axes[1].axis('off')  # Hide axes for better visualization

plt.tight_layout
plt.show()
```
